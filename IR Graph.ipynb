```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("PC+40 GLY.CSV")

df["abs"]= df["abs"]/df["abs"].max()

fig, ax= plt.subplots(1, 1)
ax.plot(df["long_onda"], df["abs"])
#cutting the plot 
ax.set_xlim(4000, 500)
ax.set_ylim(None, 1.05)
ax.set_xlabel("Longitud de onda")
ax.set_ylabel("Abs")
ax.set_title("Espectro IR")
ax.grid(True)
#fig.tight_layout()
plt.style.use('seaborn-deep')

plt.show()
```
